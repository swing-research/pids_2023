```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 100
sns.set_style("whitegrid")


```

```python
car, goat = 0, 1
things = np.array(['car', 'goat'])
yesno = np.array(['Hell no!', 'Hell yes!'])

n_games = 1000
doors_0 = np.array([car, goat, goat])
door_idx = np.arange(3)

# there are much more efficient but much less didactic ways to do this!

game_stats = np.zeros((n_games, 2))

for i_game in range(n_games):
    
    print("Host: Game number %d is on!" % (1 + i_game))
    print("Host: Which door would you like to open?")
    

    # this random permutation doesn't really change the result (why?)
    # but it makes for a "realistic" simulation
    
    randidx = np.random.permutation(3)
    doors = doors_0[randidx]

    print("\nNarrator: For the esteemed audience we reveal what's behind each door. But hush!")
    for i in door_idx:
        print("          Door number %d is hiding a %s" % (1 + i, things[doors[i]]))
        
    # you make a random choice of the door
    choice = np.random.randint(3)
    chosen_thing = doors[choice]
    print("\nYou: I choose door number %d!" % (1 + choice))
    
    # now the host opens one of the remaining two doors which hides a goat
    
    doors_not_chosen_idx = door_idx[door_idx != choice]
    
    remaining_goat_idx = doors_not_chosen_idx[doors[doors_not_chosen_idx] == goat]
    goat_door_idx = np.random.choice(remaining_goat_idx, 1)[0]
    print("\nHost: Look, let me open the door number %d. As you can see, there's a %s!!!" % 
          ((1 + goat_door_idx), things[doors[goat_door_idx]]))

    print("Host: Would you like to change your choice?\n")
    change = np.random.randint(2)
    print("You:", yesno[change])
    
    if change == True:
        final_chosen_door_idx = np.setdiff1d(door_idx, [choice, goat_door_idx])[0]
    else:
        final_chosen_door_idx = choice
    
        
    your_prize_idx = doors[final_chosen_door_idx]

    game_stats[i_game][0] = change
    game_stats[i_game][1] = your_prize_idx == car

    print("\nHost: Very well! Then you win a %s!!!\n\n" % things[your_prize_idx])
```

```python
wins_with_change = game_stats[game_stats[:, 0] == True][:, 1].sum()
n_changes = game_stats[:, 0].sum()
wins_without_change = game_stats[game_stats[:, 0] == False][:, 1].sum()
print("P(winning | changed) = %4.2f" % (100*wins_with_change / n_changes), "%")
print("P(winning | didn't change) = %4.2f" % (100*wins_without_change / (n_games - n_changes)), "%")

# Q: why doesn't it sum to zero?
# (rnd, not both choices \forall)
```
